```python
def detect_hazards( frame, conf_threshold=0.5):
    results = model(frame, conf=conf_threshold, verbose=False)
    detections = []
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
            confidence = float(box.conf[0])
            class_id = int(box.cls[0])
            detections.append({
            'bbox': [x1, y1, x2, y2],
            'confidence': confidence,
            'class': class_id,
            'timestamp': time.time()
            })
    return detections
```
